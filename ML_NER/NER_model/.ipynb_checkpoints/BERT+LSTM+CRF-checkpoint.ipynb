```python
df_data = pd.DataFrame(sentences, columns=['Sentence #', 'Word', 'Tag'])
```

```python
class SentenceGetter(object):
    
    def __init__(self, data):
        self.n_sent = 1
        self.data = data
        self.empty = False
        agg_func = lambda s: [(w, t) for w, t in zip(s["Word"].values.tolist(),
                                                           #s["POS"].values.tolist(),
                                                           s["Tag"].values.tolist())]
        self.grouped = self.data.groupby("Sentence #").apply(agg_func)
        self.sentences = [s for s in self.grouped]
    
    def get_next(self):
        try:
            s = self.grouped["Sentence: {}".format(self.n_sent)]
            self.n_sent += 1
            return s
        except:
            return None
```

```python
# Get full document data struce
getter = SentenceGetter(df_data)
```

```python
# Get sentence data
sentences = [[s[0] for s in sent] for sent in getter.sentences]
sentences[0]
```

```python
# Get tag labels data
labels = [[s[1] for s in sent] for sent in getter.sentences]
print(labels[0])
```

```python
tags_vals = list(set(df_data["Tag"].values))
```

```python
# Add X  label for word piece support
# Add [CLS] and [SEP] as BERT need
#tags_vals.append('X')
tags_vals.append('[CLS]')
tags_vals.append('[SEP]')
```

```python
tags_vals = set(tags_vals)
```

```python
tags_vals
```

```python
# Set a dict for mapping id to tag name
#tag2idx = {t: i for i, t in enumerate(tags_vals)}
tag2idx= {#'B-CERTIFICATION': 14,
 'B-FRAM': 1,
 'B-KNOW': 2,
 'B-LANG': 3,
 'B-PLAT': 4,
 'B-TOOL': 5,
 #'I-CERTIFICATION': 4,
 'I-FRAM': 6,
 'I-KNOW': 7,
 'I-LANG': 8,
 'I-PLAT': 9,
 'I-TOOL': 10,
 'O': 11,
 #'X': 6,
 '[CLS]': 12,
 '[SEP]': 0}
```

```python
tag2idx
```

```python
# Mapping index to name
tag2name={tag2idx[key] : key for key in tag2idx.keys()}
```

```python
# Len of the sentence must be not bigger than the training model
# See model's 'max_position_embeddings' = 512
max_len  = 256
```

```python
model_cp="bert-base-uncased"
```

```python
# load tokenizer
tokenizer=BertTokenizer.from_pretrained(model_cp)
```

```python
tokenized_texts = []
word_piece_labels = []
i_inc = 0
for word_list,label in (zip(sentences,labels)):
    temp_lable = []
    temp_token = []
    
    # Add [CLS] at the front 
    #temp_lable.append('[CLS]')
    #temp_token.append('[CLS]')
    
    for word,lab in zip(word_list,label):
        token_list = tokenizer.tokenize(word)
        for m,token in enumerate(token_list):
            temp_token.append(token)
            if m==0:
                temp_lable.append(lab)
            else:
              if lab != 'O':
                temp_lable.append('I-'+lab[2:])
              else:
                temp_lable.append('O')
                
    # Add [SEP] at the end
    #temp_lable.append('[SEP]')
    #temp_token.append('[SEP]')
    
    tokenized_texts.append(temp_token)
    word_piece_labels.append(temp_lable)
    
    if 5 > i_inc:
        print("No.%d,len:%d"%(i_inc,len(temp_token)))
        print("texts:%s"%(" ".join(temp_token)))
        print("No.%d,len:%d"%(i_inc,len(temp_lable)))
        print("lables:%s"%(" ".join(temp_lable)))
    i_inc +=1
    
```

```python
# Make text token into id
input_ids = pad_sequences([tokenizer.convert_tokens_to_ids(txt) for txt in tokenized_texts],
                          maxlen=max_len, dtype="long", truncating="post", padding="post")
print(input_ids[0])
```

```python
# Make label into id, pad with "O" meaning others
tags = pad_sequences([[tag2idx.get(l) for l in lab] for lab in word_piece_labels],
                     maxlen=max_len, value=tag2idx["O"], padding="post"
                     ,dtype="long", truncating="post" 
                     )
print(tags[0])
```

```python
X=input_ids
y=tags
```

```python
TAG_COUNT = 20
y = [ np.eye(TAG_COUNT)[sentence] for sentence in y]
print("X[0]:", X[0])
print("y[0]:", y[0])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1234)

print("Number of sentences in the training dataset: {}".format(len(X_train)))
print("Number of sentences in the test dataset : {}".format(len(X_test)))
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
WORD_COUNT = 30522
DENSE_EMBEDDING = 50
LSTM_UNITS = 50
LSTM_DROPOUT = 0.1
DENSE_UNITS = 100
BATCH_SIZE = 256
MAX_EPOCHS = 10
```

```python
model = Sequential()
model.add(Embedding(input_dim=WORD_COUNT, output_dim=64, input_length=256, mask_zero=False))
model.add(Bidirectional(LSTM(128, return_sequences=True)))
model.add(TimeDistributed(Dense(50, activation="relu")))
crf = CRF(20)
model.add(crf)
```

```python
model.compile(optimizer="adam", loss=crf.loss_function, metrics=[crf.accuracy])
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
```

```python
history = model.fit(X_train, y_train, batch_size = 32, epochs = 5, validation_split = 0.1, verbose = 1
                    #, callbacks=[F1score(use_char=False)]
 #                   ,callbacks=callbacks_list
                    ) 
```

```python
# Plot the graph 
plt.style.use('ggplot')

def plot_history(history):
    accuracy = history.history['crf_viterbi_accuracy']
    val_accuracy = history.history['val_crf_viterbi_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    x = range(1, len(accuracy) + 1)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(x, accuracy, 'b', label='Training acc')
    plt.plot(x, val_accuracy, 'r', label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(x, loss, 'b', label='Training loss')
    plt.plot(x, val_loss, 'r', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()

plot_history(history)
```

```python
def pred2label(pred):
    out = []
    for pred_i in pred:
        out_i = []
        for p in pred_i:
            p_i = np.argmax(p)
            out_i.append(tag2name[p_i])
        out.append(out_i)
    return out
test_pred = model.predict(X_test, verbose=1)   
pred_labels = pred2label(test_pred)
test_labels = pred2label(y_test)
```
